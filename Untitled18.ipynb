```python
import pandas
import numpy
import matplotlib.pyplot
import seaborn
import plotly.express


```

```python
df = pandas.read_csv('Sample - Superstore.csv', encoding='windows-1252')

numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit']

fig, axes = matplotlib.pyplot.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for i in range(len(numeric_columns)):
    col = numeric_columns[i]
    axes[i].hist(df[col], bins=50, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Распределение признака {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Количество')
    axes[i].grid(True, alpha=0.3)

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.figure(figsize=(10,6))
seaborn.scatterplot(data=df, x='Sales', y='Profit', hue='Category', alpha=0.6)
matplotlib.pyplot.title('Прибыль в зависимости от выручки по категориям товаров')
matplotlib.pyplot.show()

matplotlib.pyplot.figure(figsize=(10,6))
seaborn.boxplot(data=df, x='Region', y='Sales')
matplotlib.pyplot.title('Распределение выручки по регионам')
matplotlib.pyplot.show()

matplotlib.pyplot.figure(figsize=(12,6))
seaborn.barplot(data=df, x='Sub-Category', y='Profit', estimator=numpy.mean)
matplotlib.pyplot.xticks(rotation=90)
matplotlib.pyplot.title('Средняя прибыль по подкатегориям')
matplotlib.pyplot.show()
```

```python
fig1 = plotly.express.histogram(df, x='Sales', nbins=60,
                               title='Распределение выручки (Sales)',
                               marginal='box')
fig1.show()

fig2 = plotly.express.scatter(df, x='Sales', y='Profit',
                             color='Region',
                             hover_data=['Category', 'Sub-Category'],
                             title='Выручка vs Прибыль с цветовой разбивкой по регионам')
fig2.show()
```

```python
cat_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']

fig, axes = matplotlib.pyplot.subplots(3, 2, figsize=(14, 12))
axes = axes.ravel()

for i, col in enumerate(cat_cols):
    seaborn.countplot(data=df, x=col, ax=axes[i], order=df[col].value_counts().index)
    axes[i].set_title(f'Распределение признака {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Количество')
    axes[i].tick_params(axis='x', rotation=45 if col in ['Sub-Category'] else 15)

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```

```python
for col in ['Sub-Category', 'State']:
    freq = df[col].value_counts(normalize=True)
    df[f'{col}_freq'] = df[col].map(freq)

print(df[['Sub-Category', 'Sub-Category_freq']].drop_duplicates().head(10))
```

```python
sub_cat_counts = df['Sub-Category'].value_counts(normalize=True)
rare_sub_cats = sub_cat_counts[sub_cat_counts < 0.01]
if not rare_sub_cats.empty:
    print("Редкие подкатегории (менее 1%):")
    print(rare_sub_cats)
    # Агрегация прибыли по этим редким категориям
    rare_agg = df[df['Sub-Category'].isin(rare_sub_cats.index)].groupby('Sub-Category')['Profit'].agg(['mean', 'min', 'max', 'sum', 'count'])
    print("\nАгрегированные показатели прибыли для редких категорий:")
    print(rare_agg)
else:
    print("Редкие категории не обнаружены.")
```

```python
def classify_performance(row):
    if row['Profit'] > 100 and row['Discount'] < 0.2:
        return 'High_Efficiency'
    elif row['Profit'] < 0:
        return 'Loss_Making'
    else:
        return 'Standard'

df['Performance_Type'] = df.apply(classify_performance, axis=1)
print("Новая категория 'Performance_Type'")
print(df['Performance_Type'].value_counts())
```
